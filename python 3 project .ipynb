```python
year_avg_age=data.groupby("Year")["Age"].mean()
print(year_avg_age)
```

```python
sports_gender_avg_weight=data.groupby(["Sport","Sex"])["Weight"].mean()
sports_gender_avg_weight["Wrestling"]["F"]
```

```python
sport_event_count=data.groupby("Sport")["Event"].nunique().sort_values(ascending=False)
sport_event_count.plot(kind="bar")
sns.set(rc={'figure.figsize':(30,25)})
plt.title("Number of Unique  Events per sports")
plt.xlabel("Sports")
plt.ylabel("Number of Unique Events")
plt.xticks(rotation=90)
plt.show()
```

```python
year_participant_count= data.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind="bar")
sns.set(rc={'figure.figsize':(10,15)})
plt.title("Number of Participants Over the years")
plt.xlabel("Year")
plt.ylabel("Number of participant")
plt.show()
```

```python
country_avg_age=data.groupby("NOC")["Age"].mean().sort_values(ascending = False)
country_avg_age.head(10).plot(kind="bar")
plt.title("Top 10 Countries with Highest Average Age of Participnat")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.xticks(rotation =45 )
plt.show()
```

```python
country_avg_age=data.groupby("NOC")["Age"].mean().sort_values(ascending = False)
country_avg_age.tail(15).plot(kind="bar")
plt.title("Top 10 Countries with Lowest Average Age of Participnat")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.xticks(rotation =45 )
plt.show()
```

```python
sns.boxplot(data=data,x="Season",y="Age")
plt.title("Distribution of Ages by Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(data=data,x="Medal",y="Height",palette ="Set2")
plt.title("Distribution of Height by Medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
most_medals_county= data["NOC"].value_counts().idxmax()
print("Most Medals-Winning Country:",most_medals_county)
```

```python
tallest_athlete=data[data["Height"]==data["Height"].max()]
print("Tallest Athlete:")
print(tallest_athlete[["ID","Name","Height","Sport"]])
```

```python
heaviest_athlete=data[data["Weight"]== data["Weight"].max()]
print("Heaviest athlete:")
print(heaviest_athlete[["ID","Name","Weight","Sport"]])
```

```python
sns.scatterplot(data=data,x="Height",y="Weight",hue="Medal",palette="Set1")
plt.title("Athlete Height vs Weight by Medal Status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(title="Medal")
plt.show()
```

```python
medals_by_Country_year=data.pivot_table(index="NOC",columns="Year",values="Medal",aggfunc="count")
sns.heatmap(medals_by_Country_year,linewidths=0.5)
plt.title("Medals Counts by Country and Year")
plt.xlabel("Year")
plt.ylabel("Country")
plt.xticks(rotation=45)
plt.show()
```
